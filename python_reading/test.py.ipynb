```python
from snownlp import normal #normal：文字转换成拼音
from snownlp import seg #seg：中文分词
from snownlp.summary import textrank #提取文本关键词或摘要
```

```python
if __name__ == '__main__':
    t = normal.zh2hans(text) #转换简体
    sents = normal.get_sentences(t) #切分句子,按句划分，实际上划分到了逗号
    doc = []
    for sent in sents:
        words1 = seg.seg(sent) #分词
        words = normal.filter_stop(words1) #过滤（应该是提取摘要词）
        doc.append(words)          #加入字典
    rank = textrank.TextRank(doc)  #提取文本摘要
    rank.solve()
    for index in rank.top_index(5):  #提取摘要并打印前5句
        print(sents[index])
    rank1 = textrank.KeywordTextRank(doc) #提取关键词摘要
    rank1.solve()
    for w in rank1.top_index(5):
        print(w)
```
